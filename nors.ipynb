```python
import pandas as pd

# Assuming df is your DataFrame
# Replace NA in 'u_cause_description' column with "No Detail Provided"
df['u_cause_description'] = df['u_cause_description'].fillna("")
df['u_cause_description'] = df['u_cause_description'].replace("No Detail Provided", "")

```

```python
df['u_duration_hours'] = df['u_duration_hours'].fillna(0)
df['u_duration_minutes'] = df['u_duration_minutes'].fillna(0)
df['u_number_of_blocked_calls'] = df['u_number_of_blocked_calls'].fillna(0)
```

```python
neither = (df['u_city_affected'].isna()) & (df['u_states_affected'].isna())
```

```python
df = df[~neither]
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

# Ensure you have the necessary NLTK data
nltk.download('stopwords')
nltk.download('wordnet')

# Combine the text from the two columns
df['combined_text'] = df['u_incident_description'] + " " + df['u_cause_description']

# Preprocess the text
def preprocess_text(text):
    if not isinstance(text, str):
        text = str(text)
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text, re.I|re.A)
    # Tokenize text
    tokens = text.split()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

# Apply the preprocessing to each row
df['processed_text'] = df['combined_text'].apply(preprocess_text)
```

```python
# Saving the DataFrame to a CSV file
df.to_csv('report_edited.csv', index=False)
```
